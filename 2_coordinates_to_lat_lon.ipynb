```python
import pandas as pd
import re
```

```python
stations=pd.read_pickle("stations.pkl")
stations.head()
```

```python
coordinates=stations[["nördliche Breite","östliche Länge"]]
coordinates=coordinates.rename({"nördliche Breite":"lat","östliche Länge":"lon"},axis=1)
coordinates["lat"]=coordinates["lat"].str.cat(others=["N"]*17)
coordinates["lon"]=coordinates["lon"].str.cat(others=["E"]*17)
coordinates
```

```python
for x in range(len(coordinates)):
    lat=coordinates["lat"][x].replace(",",".").replace("''",'"')
    deg, minutes, seconds, direction =  re.split('[°\'"]', lat)
    coordinates["lat"][x]=(float(deg) + float(minutes)/60 + float(seconds)/(60*60)) * (-1 if direction in ['W', 'S'] else 1)
    
    lon=coordinates["lon"][x].replace(",",".").replace("''",'"')
    deg, minutes, seconds, direction =  re.split('[°\'"]', lon)
    coordinates["lon"][x]=(float(deg) + float(minutes)/60 + float(seconds)/(60*60)) * (-1 if direction in ['W', 'S'] else 1)
```

```python
coordinates
```

```python
stations=stations.merge(coordinates,left_index=True,right_index=True)
stations.drop(["nördliche Breite","östliche Länge"],axis=1,inplace=True)
```

```python
stations
```

# Rename stations

```python
i=0
for name in stations["station"]:
    splitted=name.split(' ',1)[1]
    stations["station"][i]=splitted
    i+=1
```

# Export new stations file

```python
stations.to_pickle("stations_clean.pkl")
```
